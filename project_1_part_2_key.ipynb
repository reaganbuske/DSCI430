```python
from glob import glob
import re
files = glob('./data/MinneMUDAC_raw_files/*_metro_tax_parcels.txt')
files[:3]
```

```python
YEAR_RE = re.compile(r'./data/MinneMUDAC_raw_files/(\d{4})_metro_tax_parcels.txt')
years = [int(YEAR_RE.search(f).group(1)) for f in files]
years
```

```python
years_and_files = sorted([(y, f) for y, f in zip(years, files) if y >= 2004])
years_and_files
```

```python
import pandas as pd
from dfply import *
dtypes = {'centroid_long':str, 'centroid_lat':str}
df_iter = pd.read_csv('./data/MinneMUDAC_raw_files/2015_metro_tax_parcels.txt', 
                      sep='|', 
                      chunksize=500,
                      engine='python',
                      dtype=dtypes)
first_chunk = next(df_iter)
(first_chunk
 >> select(starts_with('c'))
 >> head
)
```

```python
import pandas as pd
from csv import DictReader, Sniffer
from toolz import partition_all

all_lat_long = set([])
make_tuple = lambda d: (d['centroid_lat'], d['centroid_long'])
for y, f in years_and_files:
    print('\n\nStarting year {0} and file {1}'.format(y, f))
    with open(f) as csvfile:
        reader = DictReader(csvfile, delimiter='|')
        columns = reader.fieldnames
        for i, row in enumerate(reader):
            if i % 100000 == 0:
                print('Processing row number {0} -- length set {1}'.format(i, len(all_lat_long)))
            all_lat_long.add(make_tuple(row))
```

```python
get_chunk_set = lambda chunk: set(zip(chunk.centroid_lat, chunk.centroid_long))
get_chunk_set(first_chunk)
```

```python
dtypes = {'centroid_lat':str,'centroid_long':str}
all_lat_long = set([])
make_tuple = lambda d: (d['centroid_lat'], d['centroid_long'])
for y, f in years_and_files:
    print('\n\nStarting year {0} and file {1}'.format(y, f))
    for i, chunk in enumerate(pd.read_csv(f, chunksize=10000, sep='|', engine='python', dtype=dtypes)):
        print('Processing chunk number {0} -- length set {1}'.format(i, len(all_lat_long)))
        chunk_set = get_chunk_set(chunk)
        all_lat_long = all_lat_long.union(chunk_set)
```
